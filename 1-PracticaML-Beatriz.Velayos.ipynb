Para realizar análisis y regresión con machine learning es importante eliminar columnas que:
-No aportan valor predictivo (como IDs, URLs, descripciones largas).

-Muchos nulos.

```python
# 1. Eliminar filas con precio nulo
df_reducido = df_reducido[df_reducido['Price'].notnull()]

# 2. Imputar columnas numéricas con la mediana
num_cols = ['Bathrooms', 'Beds', 'Bedrooms']
for col in num_cols:
    df_reducido[col] = df_reducido[col].fillna(df_reducido[col].median())

# 3. Imputar columnas categóricas con 'Unknown'
cat_cols = ['Host Response Time', 'State', 'Amenities', 'City', 'Calculated host listings count','Host Total Listings Count']
for col in cat_cols:
    df_reducido[col] = df_reducido[col].fillna('Unknown')    

# 4. Eliminar columnas con más del 50% de nulos
cols_muchos_nulos = ['Square Feet', 'Monthly Price', 'Weekly Price', 'Security Deposit',
                     'Cleaning Fee', 'Has Availability', 'Host Acceptance Rate', 'Host Response Rate']
cols_existentes = [col for col in cols_muchos_nulos if col in df_reducido.columns]
df_reducido.drop(columns=cols_existentes, inplace=True)

print("Valores nulos restantes por columna:")
print(df_reducido.isna().sum().sort_values(ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Hacer una copia antes del filtrado
df_original = df.copy()

# Lista de las 16 columnas relevantes
columnas_relevantes = [
    "Host Listings Count",
    "Neighbourhood Cleansed",
    "Property Type",
    "Room Type",
    "Accommodates",
    "Bathrooms",
    "Bedrooms",
    "Beds",
    "Bed Type",
    "Amenities",
    "Price",
    "Guests Included",
    "Extra People",
    "Minimum Nights",
    "Maximum Nights",
    "Number of Reviews"  
]

# Crear un DataFrame reducido solo con las columnas relevantes
df_relevante = df[columnas_relevantes].copy()
df_original_relevante = df_relevante.copy()  # Para comparación visual

# Columnas numéricas a filtrar
columnas_filtrar = [
    "Host Listings Count",
    "Extra People",
    "Minimum Nights",
    "Maximum Nights"
]

# Función de filtrado de outliers basada en IQR
def filtrar_outliers_iqr(df, columnas):
    df_filtrado = df.copy()
    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        df_filtrado = df_filtrado[(df_filtrado[col] >= limite_inferior) & (df_filtrado[col] <= limite_superior)]
    return df_filtrado

# Aplicar el filtrado
df_reducido = filtrar_outliers_iqr(df_relevante, columnas_filtrar)

# Mostrar estadísticas antes y después
print("ANTES del filtrado de outliers:")
print(df_original_relevante[columnas_filtrar].describe())

print("\nDESPUÉS del filtrado de outliers:")
print(df_reducido[columnas_filtrar].describe())

# Visualización: Boxplots antes y después
sns.set(style="whitegrid")
for col in columnas_filtrar:
    plt.figure(figsize=(12, 5))

    # Antes
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df_original_relevante[col])
    plt.title(f'{col} - ANTES del filtrado')

    # Después
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df_reducido[col])
    plt.title(f'{col} - DESPUÉS del filtrado')

    plt.tight_layout()
    plt.show()
```

La distribución de la columna Price está fuertemente sesgada(mucha concentración de precios bajos,
y una larga cola de precios altos). 
Es conveniente aplicar una transformación logarítmica.
Como ventajas:
-Reduce el sesgo de la distribución, acercándola a una forma más normal (gaussiana).
-Mejora la estabilidad y rendimiento de modelos que asumen normalidad o varianza constante.
-Disminuye el efecto de outliers extremos.
-Puede producir predicciones más realistas y menos extremas.
